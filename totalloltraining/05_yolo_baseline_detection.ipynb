# YOLO Baseline Detection on Low-Light Images

## Objective
This notebook evaluates the performance of a YOLO object detection model on
raw low-light images without any enhancement. The results serve as a baseline
for comparison with enhanced-image detection.


```python
from ultralytics import YOLO
import os
import cv2
import matplotlib.pyplot as plt
```

```python
model = YOLO("yolov8n.pt")  # nano = fast & sufficient for baseline
```

```python
#C:\IMAGE_ENHANCEMENT_YOLO_DETECTION\TotalLoL\test\low
LOW_IMG_DIR = "../TotalLoL/test/low"
BASELINE_OUT = "../results/detection_outputs/baseline"

os.makedirs(BASELINE_OUT, exist_ok=True)
```

```python
image_files = sorted(os.listdir(LOW_IMG_DIR))

for img_name in image_files:
    img_path = os.path.join(LOW_IMG_DIR, img_name)

    results = model(
        img_path,
        conf=0.25,
        iou=0.45,
        save=True,
        project=BASELINE_OUT,
        name="raw_lowlight",
        exist_ok=True
    )
```

```python
import os
import cv2
import matplotlib.pyplot as plt

PREDICT_DIR = "runs/results/detection_outputs/baseline/raw_lowlight"

print("Images found:", os.listdir(PREDICT_DIR)[:5])

sample = os.listdir(PREDICT_DIR)[0]
img = cv2.imread(os.path.join(PREDICT_DIR, sample))
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(5,5))
plt.imshow(img)
plt.title("YOLO Baseline Detection (Raw Low-Light)")
plt.axis("off")
plt.show()
```

## Observations

- Many objects are not detected or misclassified.
- Bounding boxes are missing or inaccurate.
- Confidence scores are low.
- YOLO struggles due to poor contrast and weak edges.

These results highlight the limitations of standard object detection models
under low-light conditions.
